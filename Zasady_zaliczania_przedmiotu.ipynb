- zapoznanie z używaną w ML terminologią
- zwrócenie uwagi na pewne ważne subtelności matematyczne pojawiające się w ML

- materiał wykładany na tym przedmiocie w dużym stopniu pokrywa się z anglojęzycznymi kursami dostępnymi online - osoby bardziej zaawansowane być może chcą zrezygnować z tego przedmiotu i zapisać się od razu na kursy specjalistyczne

### Wymagania wstępne
- programowanie w języku Python
- podstawy rachunku prawdopodobieństwa
- podstawy algebry liniowej
- podstawy analizy matematycznej


### Obecność na zajęciach
- obecność na wykładzie jest nieobowiązkowa (z zastrzeżeniem ostatniego podpunktu)
  * jednakże materiał wyłożony na wykładzie obowiązuje na ćwiczeniach
  * w szczególności dotyczy to znajomości i zrozumienia wzorów matematycznych
  * w trakcie ćwiczeń NIE będzie tłumaczony materiał z wykładu - wszystkie wątpliwości dotyczące materiału należy zgłaszać na bieżąco w trakcie trwania wykładu
- na ćwiczeniach obecność jest obowiązkowa
  * student ma prawo do maksymalnie trzech nieusprawiedliwionych nieobecności - przekroczenie tego limitu skutkuje otrzymaniem oceny NZAL z ćwiczeń i niedopuszczeniem do egzaminu
  * osoby dołączające do zajęć w późniejszym terminie muszą liczyć się z faktem, że w ich przypadku dozwolona liczba nieusprawiedliwionych nieobecności na ćwiczeniach może zostać zmniejszona przez koordynatora przedmiotu
- w szczególnych wypadkach koordynator przedmiotu może zwiększyć limit nieusprawiedliwionych nieobecności na ćwiczeniach, jednakże dotyczy to tylko studentów, którzy z ćwiczeń otrzymali minimum ocenę dst
- jeśli osoby uczestniczące w ćwiczeniach notorycznie będą nieprzygotowane z materiału przedstawionego na wykładzie i będzie to uniemożliwiało sprawne prowadzenie zajęć, obecność na wykładzie stanie się dla nich obowiązkowa; w takiej sytuacji będą obowiązywały limity podobne do tych, o których mowa w wypadku ćwiczeń

## Zasady oceniania
---
### Ocena z ćwiczeń

Podczas ćwiczeń studenci otrzymają do wykonania 6 zadań programistycznych. 
- __Każde zadania oceniane jest na 0 lub 1 punkt__
- Rozwiązania zadań muszą spełniać następujące kryteria:
  - kod musi być czytelny i zwięzły (w razie wątpliwości proszę stosować się do standardu PEP 8),
  - kod musi być przetestowany i poprawnie działać,
  - jeśli do zadania dodane są asserty, kod musi je przechodzić,
  - rozwiązanie musi zostać wysłane w terminie,
  - na następnych zajęciach student może być poproszony o przedstawienie fragmentu lub całości rozwiązania i wytłumaczenie działania kodu.

Niespełnienie __któregokolwiek__ z powyższych podpunktów skutkuje otrzymaniem zera punktów za zadanie. Zadania rozwiązane z drobnymi niedociągnięciami mogą zostać warunkowo zaliczone. Zadania ocenione na zero punktów nie mogą być poprawiane w późniejszym terminie (m.in. ze względu na to, iż rozwiązania będą prezentowane na zajęciach).
Osoby dołączające do zajęć w późniejszym terminie muszą liczyć się z faktem, iż nie mogą otrzymać punktów za zaległe zadania.

Skala ocen z ćwiczeń
* 0-2 punkty: ndst
* 3 punkty: dst
* 4-5 punktów: db
* 6 punktów: bdb

Termin wystawienia ocen z ćwiczeń: 27 stycznia 2019 roku.

### Ocena końcowa

Każdy student, który zaliczył ćwiczenia (WŁĄCZNIE ze studentami, którzy otrzymali ocenę ndst) ma prawo i obowiązek przystąpić do egzaminu pisemnego. Ocena końcowa z przedmiotu (PIERWSZY TERMIN) jest oceną łączną z ćwiczeń i egzaminu, obliczaną w następujący sposób:
- za listopadowe kolokwium można otrzymać 0-20 pkt
- za egzamin można otrzymać 0-80 pkt
- za pozytywną ocenę z ćwiczeń student otrzymuje dodatkowe punkty:
    - za dst: +10 pkt
    - za db: +20 pkt
    - za bdb: +30 pkt
- łącznie można więc uzyskać od 0 do 130 pkt

Procenty przekładają się na ocenę końcową w następujący sposób:
- < 50 pkt: ndst
- < 60 pkt: dst
- < 70 pkt: dst+
- < 80 pkt: db
- < 90 pkt: db+
- \>= 90 pkt: bdb

W wypadku uzyskania w pierwszym terminie oceny niedostatecznej, student ma prawo podejść do egzaminu w DRUGIM TERMINIE. W takiej sytuacji ocena wystawiona w drugim terminie zależy wyłącznie od wyniku egzaminu poprawkowego (nie zależy od listopadowego kolokwium, nie zależy od oceny z ćwiczeń).

Powyższe zasady oceniania NIE ULEGNĄ ZMIANIE w trakcie semestru, ani na korzyść, ani na niekorzyść studentów.

Termin wystawienia ocen z przedmiotu:
- 9 lutego 2019 r. - dla osób, które zdały przedmiot w pierwszym terminie,
- 22 lutego 2019 r. - dla pozostałych.

Egzamin końcowy w pierwszym terminie przeprowadzony będzie w formie pisemnej. Będą się na niego składały pytania jednokrotnego wyboru, wielokrotnego wyboru i otwarte.

Egzamin w drugim terminie będzie przeprowadzony w formie ustnej lub pisemnej - w zależności od liczby studentów, która do niego przystąpi.

Terminy egzaminów zostaną ustalone w styczniu 2019 r.

## Dodatkowe uwagi dot. zasad oceniania

Listopadowe kolokwium ma na celu sprawdzić wiedzę i umiejętności matematyczne. Będzie składało się z kilku zadań pisemnych, podobnych do tych, które zostaną omówione na ćwiczeniach.

Natomiast pytania egzaminacyjne będzie można podzielić na trzy kategorie:
- sprawdzające znajomość wykładu (wiedza teoretyczna) - około 30-40%
- sprawdzające znajomość materiału związanego bezpośrednio z zadaniami z ćwiczeń (wiedza teoretyczno-praktyczna) - około 30-40%
- sprawdzające dogłębne zrozumienie materiału (pytania trudne i podchwytliwe) - około 10%

W związku z tym osoba, która systematycznie pracowała na ćwiczeniach przez cały rok najprawdopodobniej nawet bez przygotowania do egzaminu otrzyma ocenę końcową na poziomie db (przykładowo: 30 pkt z ćwiczeń + około 40 pkt z egzaminu + około 10 pkt z kolokwium). 

Natomiast osoba, która nie otrzymała oceny pozytywnej z ćwiczeń najprawdopodobniej otrzyma ocenę końcową na poziomie dst pod warunkiem, że bardzo solidnie zrozumie materiał z wykładu i ćwiczeń. Praktyka pokazuje jednak, że aby zdać niniejszy przedmiot, najbezpieczniej mieć przynajmniej ocenę db z ćwiczeń.


## Samodzielna praca

Celem przedmiotu jest przede wszystkim przygotowanie studentów do dalszej działalności w dziedzinie Machine Learning. W związku z tym zachęcamy studentów do:
- (oczywiście) aktywnego uczestnictwa w zajęciach,
- korzystania z wyznaczonych przez nas terminów konsultacji,
- dyskutowania między sobą na temat wykładów i ćwiczeń oraz zadawanych zadań,
- odwiedzania Koła Naukowego Robotyki i Sztucznej Inteligencji.

Jednakże aby ocena z przedmiotu wystawiona została sprawiedliwie, część pracy musi zostać wykonana samodzielnie przez studenta. Aby uniknąć niepotrzebnych nieporozumień, proszę bardzo dokładnie zapoznać się z poniższymi zasadami.

1. Dyskusja na temat zadawanych zadań może dotyczyć jedynie ogólnej postaci algorytmu. Prosimy w miarę możliwości zadawać pytania dotyczące zadań na ćwiczeniach.
2. Jeśli studenci dyskutują między sobą (w mniejszej grupie) na temat zadań i tematyka dyskusji istotnie przekracza materiał wyłożony na wykładzie lub ćwiczeniach, student ma obowiązek zamieścić krótkie podsumowanie dyskusji na forum Pegaz.
3. Niedozwolone jest:
    - wspólne rozwiązywanie zadań programistycznych,
    - udostępnianie innym studentom rozwiązania zadania w formie szczegółowego (krok po kroku) opisu algorytmu lub kodu,
    - posługiwanie się notatkami w trakcie trwania egzaminu końcowego, "ściąganie".

Każdy student, który stwierdzi naruszenie powyższych zasad, ma obowiązek poinformować o tym koordynatora przedmiotu.

W zależności od skali przewinienia kary za złamanie powyższych zasad to:
- bezwarunkowe obniżenie oceny końcowej z przedmiotu,
- bezwarunkowe NZAL z całego przedmiotu,
- postępowanie dyscyplinarne.
